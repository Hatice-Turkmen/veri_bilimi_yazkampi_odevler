```python
"""Soru 1 – Liste Metotları
Bir sınıfta öğrencilerin notları şu şekilde tutuluyor:
notlar = [85, 92, 76, 92, 100, 76, 85, 92]
•
Listedeki tekrar eden notları silip benzersiz bir liste oluşturun.
•
En yüksek ve en düşük notu bulun.
•
Notları küçükten büyüğe sıralayın.
"""
notlar = [85, 92, 76, 92, 100, 76, 85, 92]
yeni_list=list(set(notlar))
en_yuksek=max(yeni_list)
en_dusuk=min(yeni_list)
sirali_notlar = sorted(yeni_list)
print(f"Yeni liste:{yeni_list}\n"
      f"Max not:{en_yuksek}\n"
      f"Min not:{en_dusuk}\n"
      f"Yeni liste:{yeni_list}\n")
```

```python
"""Soru 2 – Sayılar
Bir sayının Armstrong sayısı olup olmadığını kontrol eden bir Python fonksiyonu yazın.
Armstrong sayısı: Her basamağının küplerinin toplamı kendisine eşit olan sayılar. Örn: 153 → 1³+5³+3³ = 153
"""
sayi=int(input("sayı giriniz:"))
basamak=str(sayi)
toplam=0
for i in basamak:
    arms=int(i)**len(basamak)
    toplam+=arms
if (sayi==toplam):
    print("Sayı Armstrong sayısıdır.")
else:
    print("Sayı Armstrong sayısı değildir.")
```

```python
"""Soru 3 – Kümeler
Aşağıdaki iki küme verilmiştir:
A = {"Python", "R", "SQL", "Java"}
B = {"C++", "Python", "JavaScript", "SQL"}
•
Ortak dilleri bulun.
•
Sadece A’da olan dilleri listeleyin.
•
İki kümenin birleşimini alfabetik olarak yazdırın."""
A = {"Python", "R", "SQL", "Java"}
B = {"C++", "Python", "JavaScript", "SQL"}
ortak=A&B
sadece_A=A-B
birlesim = sorted(A | B) #sorted= alfabetik veya sayısal olarak sıralamak için, A|B de küme birleşimi
print(f"Ortak olanlar:{ortak}\n"
     f"Sadece A'da olanlar:{sadece_A}\n"
     f"Birleşimi:{birlesim}")
```

```python
"""Soru 4 – Modüller
•
random modülünü kullanarak 1–100 arasında 10 rastgele sayı üretin.
•
Bu sayıların ortalamasını ve standart sapmasını statistics modülü ile hesaplayın.
"""
import random
import statistics
sayilar = [random.randint(1, 100) for _ in range(10)]
ortalama = statistics.mean(sayilar)
standart_sapma = statistics.stdev(sayilar)
print(f"Rastgele Sayılar: {sayilar}\n"
     f"Sayıların ortalaması: {ortalama}\n"
     f"Sayıların standart sapması: {standart_sapma}")
```

```python
"""Soru 5 – Fonksiyonlar
kelime_sayacı(metin) adında bir fonksiyon yazın. Fonksiyon verilen metindeki:
•
toplam kelime sayısını
•
en uzun kelimeyi
•
en sık geçen kelimeyi döndürsün.
"""
from collections import Counter  #collections modülünde bulunan Counter sınıfı, bir koleksiyondaki elemanların kaç kez tekrar ettiğini saymak için kullanılır

def kelime_sayaci(metin):
    kelimeler= metin.split()
    toplam = len(kelimeler)
    en_uzun = max(kelimeler, key=len)  #max() fonksiyonu normalde elemanları alfabetik olarak karşılaştırırdı ama bir listeyi sıralarken sadece elemanların kendisini değil uzunluk, kare gibi başka özelliğini de kullanmak istiyorsan key kullanırsın.
    en_sık = Counter(kelimeler).most_common(1)[0][0]   #most_common(1) → en sık tekrar eden 1 elemanı verir. [0][0] → önce tuple’ı, sonra elemanı alır.
    
    return toplam, en_uzun, en_sık

metin ="Ah şu anda 3.haftanın ödevini yapıyorum ve 3.haftanın ödevide önceki haftaların ödevi gibi gidiyor şimdilik. "
toplam, en_uzun, en_sık = kelime_sayaci(metin)
print("Toplam kelime sayısı:", toplam)
print("En uzun kelime:", en_uzun)
print("En sık geçen kelime:", en_sık)
```

```python
"""Soru 7 – Lambda İfadeleri
Aşağıdaki listeyi, her kelimenin uzunluğuna göre küçükten büyüğe sıralayın.
kelimeler = ["veri", "bilim", "analiz", "yapayzeka", "python"]
Bunu sorted + lambda ile yapın.
"""
kelimeler = ["veri", "bilim", "analiz", "yapayzeka", "python"]
sirali = sorted(kelimeler, key=lambda kelime: len(kelime))    #lambda ile geçici bir fonksiyon tanımlanarak, her eleman için karşılaştırma kriteri belirlenir
print(sirali)
```

```python
"""Soru 8 – Metodlar
Bir string içinde geçen tüm rakamları bulun ve bunların toplamını döndüren bir fonksiyon yazın. Örn: "abc12def3" → 12 + 3 = 15
"""
def rakam_toplami(metin):
    toplam = 0
    for karakter in metin:
        if karakter.isdigit():   # rakam mı?
            toplam += int(karakter)
    return toplam

metin = "abc12def3"
print(rakam_toplami(metin))
```

```python
"""Proje – “Kitap Satış Analiz Sistemi”"""
import random
import statistics
from collections import Counter

kitaplar = [
{"isim": "Veri Bilimi 101", "yazar": "Ali", "tur": "Bilim", "satis": 1200, "yil": 2021},
{"isim": "Python ile Yapay Zeka", "yazar": "Ayşe", "tur": "Bilim", "satis": 950, "yil": 2020},
{"isim": "İstatistik Temelleri", "yazar": "Ali", "tur": "Akademik", "satis": 700, "yil": 2019},
{"isim": "Makine Öğrenmesi", "yazar": "Can", "tur": "Bilim", "satis": 1800, "yil": 2022},
{"isim": "Veri Görselleştirme", "yazar": "Deniz", "tur": "Sanat", "satis": 400, "yil": 2018},
{"isim": "Matematiksel Modelleme", "yazar": "Ali", "tur": "Akademik", "satis": 1500, "yil": 2021},
{"isim": "Bilgi Toplumu", "yazar": "Ayşe", "tur": "Sosyal", "satis": 600, "yil": 2022}
]
def en_cok_satan(kitaplar):
    return max(kitaplar, key=lambda k: k["satis"])
def yazar_satislari(kitaplar):
    satis_dict = {}
    for kitap in kitaplar:
        satis_dict[kitap["yazar"]] = satis_dict.get(kitap["yazar"], 0) + kitap["satis"]
    return satis_dict

turler = set(kitap["tur"] for kitap in kitaplar)
binden_fazla_satanlar = [kitap["isim"] for kitap in kitaplar if kitap["satis"] > 1000]
sonrasi_2020 = list(filter(lambda k: k["yil"] > 2020, kitaplar))
satis_artir = list(map(lambda k: k["satis"]*1.1, kitaplar))
sirali_satis = sorted(kitaplar, key=lambda k: k["satis"], reverse=True)
satislar = [kitap["satis"] for kitap in kitaplar]
ortalama_satis = statistics.mean(satislar)
std_sapma = statistics.stdev(satislar)
tur_satislari = {}
for kitap in kitaplar:
    tur_satislari[kitap["tur"]] = tur_satislari.get(kitap["tur"], 0) + kitap["satis"]
en_cok_satan_tur = max(tur_satislari, key=tur_satislari.get)

train_size = int(0.7 * len(kitaplar))
train_set = random.sample(kitaplar, train_size)
test_set = [k for k in kitaplar if k not in train_set]
train_yazar_ortalama = {yazar: statistics.mean([k["satis"] for k in train_set if k["yazar"]==yazar])
                        for yazar in set(k["yazar"] for k in train_set)}

test_ustu = [k["isim"] for k in test_set if k["satis"] > max(train_yazar_ortalama.values())]


print(f"""
En çok satan kitap: {en_cok_satan(kitaplar)['isim']}
Yazar satışları: {yazar_satislari(kitaplar)}
Tüm türler: {turler}
1000'den fazla satan kitaplar: {binden_fazla_satanlar}
2020 sonrası çıkan kitaplar: {[k['isim'] for k in sonrasi_2020]}
Satışları %10 artırılmış satışlar: {satis_artir}
Kitaplar satışa göre azalan sıralama: {[k['isim'] for k in sirali_satis]}
Ortalama satış: {ortalama_satis:.2f}
Standart sapma: {std_sapma:.2f}
En çok satış yapan tür: {en_cok_satan_tur}
Train set yazar ortalamaları: {train_yazar_ortalama}
Test setinde ortalamanın üzerinde satan kitaplar: {test_ustu}
""")
```
